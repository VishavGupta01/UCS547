```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])

def function(arr, name):
    vals, counts = np.unique(arr, return_counts=True)
    max_count = counts.max()

    modes = vals[counts == max_count]

    print(f"Results for array {name}:")
    for mode in modes:
        indices = np.where(arr == mode)[0]
        print(f"Value: {mode}, Indices: {indices}")
    print("")

function(x, "x")
function(y, "y")
```
